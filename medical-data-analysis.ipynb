## Genism word2vec Model Training

We can train the genism word2vec model with our own custom corpus as following:

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import gensim
```

```python
from gensim.models import Word2Vec
```

Explain model:
- size: The number of dimensions of the embeddings and the default is 100.
- window: The maximum distance between a target word and words around the target word. The default window is 5. Since our text is shorter, we set window as 3.
- min_count: The minimum count of words to consider when training the model; words with occurrence less than this count will be ignored. The default for min_count is 5.
- workers: The number of partitions during training and the default workers is 3.
- sg: The training algorithm, either CBOW(0) or skip gram(1). The default training algorithm is CBOW. We use skip gram here. 

Train the model:

```python
model = Word2Vec(sent, min_count=1,size= 50,workers=3, window =3, sg = 1)
```

Since each unique hospital visit for a patient is assigned a unique HADM_ID, the ADMISSIONS table can be considered as a definition table for HADM_ID. It is possible for this table to have duplicate SUBJECT_ID, indicating that a single patient had multiple admissions to the hospital. However, for each hospital visit, there could be multiple treatments. Thus, there are multiple rows for one HADM_ID. We want to map each hospital visit (each unique HADM_ID) to a list of records, which can be seen as a event-matrix.

use numpy.mean to center the matrix.

```python
event_seq = dict()
for row in l:
    hadm_id = row['HADM_ID']
    if hadm_id not in event_seq.keys():
        event_seq[hadm_id] = list()
        event_seq[hadm_id].append(numpy.mean(row['RECORD_VECTOR']))
    else:
        event_seq[hadm_id].append(numpy.mean(row['RECORD_VECTOR']))
```
